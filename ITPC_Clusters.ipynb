```python
# Set up OpenAI API key
openai.api_key = 'key'

# Function to classify text content using GPT-3.5 prompting
def classify_with_gpt(text):
    
    # Define prompt instruction for all IPTC categories
    prompt_instruction = "Please classify the following news under the appropriate category:"
    
    # Define prompts for each IPTC category
    prompts = {
        'Politics': "Politics:",
        'Business': "Business:",
        'Sports': "Sports:",
        'Entertainment': "Entertainment:",
        'Technology/Science': "Technology/Science:",
        'Health/Medical': "Health/Medical:",
        'Environment': "Environment:",
        'Crime/Law': "Crime/Law:",
        'Human Interest': "Human Interest:",
        'Travel': "Travel:",
        'Education': "Education:"
    }
    
    # Combine prompts
    combined_prompt = f"{prompt_instruction}\n\n{text}\n"
    for category, category_prompt in prompts.items():
        combined_prompt += f"{category_prompt} Category: {category}\n"
    
    # Generate text using GPT-3.5 with prompting
    response = openai.Completion.create(
        engine="gpt-3.5-turbo-instruct",
        #engine="gpt-3.5-turbo-0125",
        prompt=combined_prompt,
        max_tokens=100,
        temperature=0.5,
        top_p=1.0,
        n=1,
        stop=None
    )
    generated_text = response.choices[0].text.strip()
    
    # Extract category from generated text
    for category in prompts.keys():
        if category.lower() in generated_text.lower():
            return category
    
    # If no category is indicated, return None
    return None

# Apply classification function to each row and add IPTC group to dataframe
df3['IPTC group'] = df3['Text'].apply(classify_with_gpt)

# Display 
df3
```

```python
import pandas as pd

# Assuming keys_df is your DataFrame

# Identify unique values in the 'IPTC' column
unique_values = filtered_dff['clusters'].unique()

# List to store filtered DataFrames
filtered_dfs = []

# Iterate through unique values
for value in unique_values:
    # Filter DataFrame to include only rows with the current value
    temp_df = filtered_dff[filtered_dff['clusters'] == value]
    # If there's only one row for the value or if the value is unique, append it to filtered_dfs
    if len(temp_df) == 1 or len(filtered_dff[filtered_dff['clusters'] == value]) == 1:
        filtered_dfs.append(temp_df)
    else:
        # If there are multiple rows with the same value, append only the first row
        filtered_dfs.append(temp_df.iloc[:1])

# Concatenate filtered DataFrames into one DataFrame
filtered_df3 = pd.concat(filtered_dfs)

# Print the filtered DataFrame
filtered_df3
```
